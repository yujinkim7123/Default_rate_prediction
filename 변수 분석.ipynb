 심리지수 변수 선정
부도확률 예측에 유용한 심리지수 관련 변수는 기업의 경제적 신뢰도와 향후 경제 환경에 대한 기대감을 반영하기 때문에, 심리지수의 중요도를 우선순위로 매기려면 기업 활동에 미치는 영향과 경제 불확실성을 기준으로 정리

우선순위 선정
기업경기실사지수 (BSI, Business Survey Index)
우선순위 1: 기업경기실사지수(BSI)는 기업들이 느끼는 현재 경기 상황과 향후 경기 전망을 반영하는 지표
기업들의 경영환경에 대한 직접적인 심리를 나타내기 때문에 부도확률 예측에 가장 중요한 변수 중 하나
주요 항목: 현재 경기 상황, 향후 경기 전망, 재고 수준, 수출 및 내수 전망 등.
부도예측 이유: 기업들의 신뢰도가 떨어지면 생산과 투자 계획이 위축되고, 자금 흐름이 악화되어 부도의 위험이 높아짐

2. 소비자 심리지수 (CSI, Consumer Sentiment Index)
우선순위 2: 소비자 신뢰지수(CSI)는 소비자들이 느끼는 경제 상황과 향후 경제 전망에 대한 기대감을 반영합니다. 소비자 신뢰도가 낮으면 소비가 위축되어 기업들의 매출이 감소하고, 이는 부도의 위험을 증가시킬 수 있습니다.
주요 항목: 현재 경제 상황 평가, 향후 경제 상황 기대, 소비지출 예상, 고용 전망 등.
부도예측 이유: 소비자 신뢰도가 낮으면 기업들의 수익성에 부정적인 영향을 미치며, 결국 부도의 위험이 커집니다.

3. 금융시장 심리지수 (특히 주식시장 변동성)
우선순위 3: 금융시장의 변동성 지수(VIX)나 주식시장 변동성은 투자자들의 불안감을 반영합니다. 시장의 불확실성이 커지면 자금 조달이 어려워지고, 기업의 신뢰도가 떨어져 부도 위험이 증가할 수 있습니다.
주요 항목: 주식시장 변동성, 신용스프레드, CDS 프리미엄 등.
부도예측 이유: 금융시장 불안은 기업들이 자금을 조달하거나 신용을 유지하는 데 큰 영향을 미칩니다.

4. 경제정책 불확실성 지수 (Economic Policy Uncertainty Index)
우선순위 4: 정부의 경제 정책 변화나 정치적 불확실성에 대한 경제 주체들의 반응을 반영하는 지표입니다. 정책 불확실성이 커지면 기업들의 투자 및 경영 전략이 위축되며, 이는 부도 확률을 높일 수 있습니다.
주요 항목: 정책 변화에 대한 불확실성, 금리 변화 전망, 규제 변경 등.
부도예측 이유: 불확실한 정책 환경은 기업들이 미래의 경제 상황을 예측하기 어렵게 만들어 경영에 큰 부담을 줍니다.

5. 기대 인플레이션율
우선순위 5: 기대 인플레이션율은 경제 주체들이 예상하는 물가 상승률로, 물가 상승에 대한 불확실성이 커질 경우 기업들은 경영 계획을 변경하고, 자금 조달이나 가격 책정에 어려움을 겪을 수 있습니다.
주요 항목: 소비자 및 기업들의 기대 인플레이션율.
부도예측 이유: 높은 기대 인플레이션은 비용 상승과 금리 인상을 유발하여 기업들의 수익성 및 유동성에 부정적 영향을 미칩니다.


6. 소비자 물가 지수 (CPI)와 경제 전반의 불안정성
우선순위 6: 물가 상승률(CPI)은 경제 전반의 인플레이션 수준을 나타냅니다. 물가 상승은 소비자와 기업의 경제적 부담을 증가시킬 수 있으며, 이에 따라 경기가 위축되면 부도 확률이 높아집니다.
주요 항목: 소비자 물가 지수, 물가 상승률.
부도예측 이유: 물가 상승은 소비 위축과 함께 기업들의 비용 증가로 이어져 부도의 위험을 키울 수 있습니다.

7. 국제 경제 불확실성 (Global Economic Uncertainty)
우선순위 7: 국제 경제 불확실성은 국내 기업들에게 직접적인 영향을 미칩니다. 글로벌 경제 위기나 불확실성이 커지면 기업들이 직면하는 위험도 높아집니다.
주요 항목: 글로벌 경기 침체, 국제 유가 변동, 국제 금융 위기 등.
부도예측 이유: 해외 시장에 의존하는 기업들은 글로벌 경제 불안정성이 클 경우, 자금 조달과 수출에 큰 어려움을 겪게 됩니다.


종합적으로,
부도확률 예측을 위한 심리지수 관련 변수들은 기업과 소비자들이 현재와 미래의 경제 환경을 어떻게 인식하고 있는지 반영하는 지표입니다. 기업경기실사지수와 소비자 심리지수는 기업과 소비자 신뢰를 직접적으로 보여주는 지표로 가장 중요한 변수가 되며, 금융시장 변동성이나 경제정책 불확실성은 기업의 경영 및 자금 조달에 큰 영향을 미치는 요소입니다. 이러한 지표들을 종합적으로 분석함으로써 부도 위험을 더 정확하게 예측할 수 있습니다

부도예측을 위한 종합 점수를 계산할 때 각 항목의 가중치를 조정하는 것은 매우 중요합니다. 가중치를 어떻게 설정할지에 대한 전략을 세우기 전에, 각 항목이 부도예측에 미치는 영향을 잘 반영할 수 있도록 가중치를 조정하는 방법에 대해 좀 더 구체적으로 설명드리겠습니다.
부도예측을 잘 하기 위해서는 경제적 건강 상태, 소비 및 저축 상태, 부채 수준등이 중요한 역할을 합니다. 부도예측 모델에서 중요한 지표를 파악하고, 이를 반영할 수 있도록 가중치를 조정해야 합니다.
부도예측에 유의미한 항목들
1. 현재생활형편CSI: 현재 생활 수준이 나쁘면 채무불이행 위험이 높아질 수 있습니다.
2. 현재경기판단CSI: 경제 상황을 얼마나 잘 인식하는지는 기업이나 개인의 채무불이행과 연관이 있을 수 있습니다.
3. 생활형편전망CSI: 미래의 생활 수준에 대한 전망도 중요한 지표입니다.
4. 향후경기전망CSI: 경제 전망이 부정적일 경우 채무불이행 확률이 높아질 수 있습니다.
5. 취업기회전망CSI: 고용 상황은 부도 예측에 중요한 영향을 미칩니다.
6. 금리수준전망CSI: 금리가 높을 경우 채무 부담이 커져 부도 가능성이 높아질 수 있습니다.
7. 가계수입전망CSI: 소득이 안정적이지 않으면 부도의 위험이 커집니다.
8. 소비지출전망CSI: 과도한 소비 지출은 채무 불이행 위험을 높일 수 있습니다.
9. 현재가계저축CSI: 저축이 적다면 긴급 상황에서 부도가 날 가능성이 높습니다.
10. 가계저축전망CSI: 미래 저축 전망도 부도예측에 중요한 요소입니다.
11. 현재가계부채CSI: 높은 부채는 직접적으로 부도의 위험을 증가시킵니다.
이 항목들이 부도예측에서 중요한 요소이기 때문에, 각 항목의 가중치를 부도예측의 가능성을 잘 반영할 수 있도록 조정해야 합니다.
부도예측에 유리한 가중치 예시
1. 현재가계부채CSI: 부채는 부도의 가장 중요한 지표이므로 가장 높은 가중치를 부여합니다.
2. 현재생활형편CSI: 현재 생활 상태가 부도 예측에 중요한 영향을 미치므로 높은 가중치를 부여합니다.
3. 생활형편전망CSI: 미래 생활 수준의 변화도 부도 가능성과 연관이 있으므로 중간 정도의 가중치를 부여합니다.
4. 향후경기전망CSI: 경기 전망이 나쁘다면 부도 위험도 증가할 수 있으므로 중요한 항목으로, 중간 정도의 가중치를 부여합니다.
5. 취업기회전망CSI: 고용 불안정은 부도의 위험을 증가시키므로 중간 정도의 가중치를 부여합니다.
6. 가계수입전망CSI: 소득 감소나 불안정은 부도 가능성을 높이므로 중간 정도의 가중치를 부여합니다.
7. 금리수준전망CSI: 금리가 높을수록 부채 부담이 커지므로 중간 정도의 가중치를 부여합니다.
8. 소비지출전망CSI: 과도한 소비는 부도를 일으킬 수 있으므로 상대적으로 중요한 항목으로 간주하여 중간 정도의 가중치를 부여합니다.
9. 가계저축전망CSI: 저축이 부족하면 긴급 상황에서 부도가 날 가능성이 높기 때문에 중요한 항목으로 설정합니다.
예시 가중치
• 현재가계부채CSI: 0.3
• 현재생활형편CSI: 0.2
• 생활형편전망CSI: 0.15
• 향후경기전망CSI: 0.1
• 취업기회전망CSI: 0.1
• 가계수입전망CSI: 0.1
• 금리수준전망CSI: 0.05
• 소비지출전망CSI: 0.05
• 가계저축전망CSI: 0.05
위의 가중치는 각 항목의 중요도에 맞춰 조정된 예시입니다. 실제 모델링을 할 때는 데이터 분석을 통해 각 항목이 부도 예측에 미치는 영향을 확인하고 이를 바탕으로 가중치를 조정해야 할 수 있습니다.


```python
import pandas as pd
import re
# 1. CSV 파일 불러오기 (예시로 'csi_data.csv'로 가정)
df = pd.read_csv('/content/drive/MyDrive/소비자동향조사(전국, 월, 2008.9~)_23101152.csv')

# 2. 필요한 열만 추출 (연도/월 데이터 열)
columns_to_keep = ['CSI코드', 'CSI분류코드']
colums_to_number = [str(year) + '/' + str(month).zfill(2) for year in range(2009, 2024) for month in range(1, 13)]
colums_to_total = ['CSI코드', 'CSI분류코드'] + colums_to_number
df = df[colums_to_total]

# 쉼표 및 공백 제거
df = df.applymap(lambda x: x.replace('·', '') if isinstance(x, str) else x)
df = df.applymap(lambda x: x.replace(' ', '') if isinstance(x, str) else x)

# 3. CSI 항목과 가중치 설정
categories = {
    '경제인식조사': {
        '항목': ['현재생활형편CSI', '현재경기판단CSI'],
        '항목별가중치': {'현재생활형편CSI': 0.2 / 0.35, '현재경기판단CSI': 0.15 / 0.35}
    },
    '경제전망조사': {
        '항목': ['생활형편전망CSI', '향후경기전망CSI', '취업기회전망CSI', '금리수준전망CSI'],
        '항목별가중치': {
            '생활형편전망CSI': 0.15 / 0.4,
            '향후경기전망CSI': 0.1 / 0.4,
            '취업기회전망CSI': 0.1 / 0.4,
            '금리수준전망CSI': 0.05 / 0.4
        }
    },
    '소비지출전망조사': {
        '항목': ['가계수입전망CSI', '소비지출전망CSI'],
        '항목별가중치': {'가계수입전망CSI': 0.1 / 0.15, '소비지출전망CSI': 0.05 / 0.15}
    },
    '가계저축 및 부채조사': {
        '항목': ['현재가계저축CSI', '가계저축전망CSI', '현재가계부채CSI'],
        '항목별가중치': {'현재가계저축CSI': 0.05 / 0.4, '가계저축전망CSI': 0.05 / 0.4, '현재가계부채CSI': 0.3 / 0.4}
    }
}

# 4. CSI 데이터를 피벗하여 다룬다.
df_csi = df

#print(df_csi)

# 5. 연도별 종합 점수 계산
def calculate_weighted_score(df_csi, category, year):
    weighted_sum = 0

    # 각 CSI 항목에 대한 가중치 적용
    for item, weight in categories[category]['항목별가중치'].items():
        # 연도별 월별 데이터를 선택
        monthly_columns = [col for col in df_csi.columns if col.startswith(year)]  # 연도별 월별 데이터를 선택
        df_select = df_csi[df_csi['CSI코드']==item]

        monthly_values = df_select[monthly_columns].values

        # 월별 값의 평균을 구해 가중치를 곱함
        weighted_sum += monthly_values.mean() * weight

    return weighted_sum

# 6. 모든 년도별 종합지표 계산 (연도별)
def compute_combined_index_for_all_years(df_csi):
    combined_index = {}
    years = [str(year) for year in range(2009, 2024)]  # 2008~2023년

    # 각 연도별로 종합 점수 계산
    for year in years:
        category_scores = {}
        # 각 카테고리별 종합 점수 계산
        for category in categories:
            category_scores[category] = calculate_weighted_score(df_csi, category, year)

        combined_index[year] = category_scores

    return combined_index

# 7. 결과 출력 및 CSV 파일로 저장
combined_index = compute_combined_index_for_all_years(df_csi)


# DataFrame으로 변환하여 CSV로 저장
final_result = pd.DataFrame(combined_index).T  # Transpose로 연도별로 종합 점수가 나오도록

# 파일로 저장
final_result.to_csv('/content/drive/MyDrive/combined_index_by_category.csv', encoding='utf-8-sig')

print("CSV 파일로 저장 완료!")
```

부도 확률 예측에 적합한 변수의 가중치를 설정하려면, 업황 실적, 매출 실적, 자금 사정 실적, 인력 사정 실적 등의 항목을 기반으로 중요한 변수를 설정해야 합니다. 각 항목의 중요도에 따라 가중치를 조정하여 부도 확률 예측에 유용한 모델을 구축할 수 있습니다.

부도 확률 예측은 경제적 불확실성, 기업의 재무 건전성, 시장 상황 등을 기반으로 하기 때문에, 각각의 항목이 부도와 관련된 중요한 신호일 수 있습니다. 일반적으로, 자금 사정 실적과 매출 실적이 부도 확률에 더 큰 영향을 미칠 수 있습니다. 이에 따라 각 항목에 대해 적절한 가중치를 설정하는 것이 중요합니다.

변수별 가중치 제안:
업황 실적 (업황실적BSI): 기업이 속한 업종의 경제적 상황과 전반적인 업황을 반영합니다. 부도와의 관계는 간접적일 수 있지만, 업황이 좋지 않으면 기업이 경영하기 어려운 상황에 처할 수 있습니다.
가중치 제안: 0.25 (업황 실적이 부도에 영향을 미칠 수 있지만, 다소 간접적)

매출 실적 (매출실적BSI): 매출은 기업의 핵심적인 재무 성과 지표입니다. 부도와 직결되는 경우가 많습니다. 매출이 감소하면 현금 흐름에 어려움이 생기고, 이는 부도 위험을 높입니다.
가중치 제안: 0.30 (매출 실적은 부도 확률에 매우 중요한 영향을 미칠 수 있음)

자금 사정 실적 (자금사정실적BSI): 자금 사정은 기업의 유동성과 직결되며, 유동성이 부족하면 기업은 채무 상환을 못 하거나 경영에 큰 어려움을 겪을 수 있습니다.
가중치 제안: 0.35 (자금 사정은 부도 확률에 매우 중요한 요소)

인력 사정 실적 (인력사정실적BSI): 인력 사정은 기업의 생산성 및 인적 자원 관리 상태를 나타냅니다. 경영난이나 부진한 경제 상황에서는 인력 감소, 해고, 인력 부족 등이 발생할 수 있으며, 이는 간접적으로 부도 확률에 영향을 미칠 수 있습니다.
가중치 제안: 0.10 (인력 사정은 중요한 지표일 수 있지만, 매출과 자금 사정보다는 상대적으로 낮은 중요도를 가질 수 있음)

최종 가중치:
업황실적BSI: 0.25
매출실적BSI: 0.30
자금사정실적BSI: 0.35
인력사정실적BSI: 0.10
이제 이 가중치를 바탕으로 가중 평균을 계산하여 부도 확률 예측에 유용한 종합 점수를 도출할 수 있습니다. 가중치를 설정한 후, 각 항목에 대해 가중치를 곱하고 합산하는 방식으로 계산하면 됩니다.

```python
import pandas as pd
import re  # 정규 표현식 모듈

# 예시 CSV 파일 경로 (실제 CSV 파일 경로를 입력하세요)
file_path = '/content/drive/MyDrive/기업경기조사(실적)_23022123.csv'

# CSV 파일에서 데이터를 읽어오기
df = pd.read_csv(file_path, encoding='utf-8')

# 불필요한 열 제거
df = df.drop(['통계표', '단위', '변환'], axis=1)

df.fillna(0, inplace=True)

# BSI코드 열에서 항목명만 추출 (숫자와 괄호 제거)
def extract_bsi_name(bsi_code):
    # 정규식: 숫자와 괄호를 제거하고, 항목명만 추출
    match = re.match(r'([^\d\(\)]+)', bsi_code)
    return match.group(1).strip() if match else bsi_code.strip()

# BSI코드 열에서 항목명만 추출하여 새로운 열로 저장
df['BSI항목'] = df['BSI코드'].apply(extract_bsi_name)

importance_weights = {
    '업황실적BSI': 0.25,
    '매출실적BSI': 0.30,
    '자금사정실적BSI': 0.35,
    '인력사정실적BSI': 0.10
}

# 날짜 열 목록 추출 (열 이름에 연도/월이 포함된 열만 추출)
date_columns = [col for col in df.columns if re.match(r'\d{4}/\d{2}', col)]

# 연도별 종합 점수 계산 함수
def weighted_average(row):
    weighted_values = 0

    # 연도별 데이터를 처리
    yearly_data = {}

    for col in date_columns:
        year = col.split('/')[0]  # '2008/01'에서 '2008'만 추출
        bsi_name = row['BSI항목']
        weight = importance_weights.get(bsi_name, 0)  # 기본 가중치는 0

        if year not in yearly_data:
            yearly_data[year] = []

        yearly_data[year].append(float(row[col]) * weight)

    # 각 연도별로 평균 계산
    for year, values in yearly_data.items():
        weighted_values += sum(values) / len(values)

    return weighted_values

# 종합 변수 계산 (모든 연도에 대해 자동으로 계산)
df['종합 변수'] = df.apply(weighted_average, axis=1)

# 결과 출력 (업종별로 종합 변수 확인)
#print(df[['업종코드', '종합 변수']])

# CSV 파일로 저장 (결과를 새로운 CSV 파일로 저장)
df.to_csv('업종별_종합_결과.csv', index=False)

# 종합 지표를 연도별로 평균 계산하여 CSV 파일로 저장할 수도 있음
df_melted = pd.melt(df, id_vars=['업종코드', 'BSI항목'], value_vars=date_columns, var_name='년도', value_name='BSI')

# 연도별 평균 종합 지수 계산
df_melted['연도'] = df_melted['년도'].apply(lambda x: x.split('/')[0])  # 연도만 추출
annual_summary = df_melted.groupby('연도')['BSI'].mean().reset_index()

# 연도별 평균 종합 지표 CSV로 저장
annual_summary.to_csv('연도별_종합_지표.csv', index=False)
```

부도를 예측하기 위해 제공된 **계정 항목**들을 활용한 새로운 변수를 생성하려면, 각 항목이 기업의 재무 상태와 부도 확률에 어떻게 영향을 미치는지를 고려해야 합니다. 부도 예측 모델에서는 **재무 건전성**, **자산-부채 구조**, **수익성** 등을 평가하는 다양한 비율들이 중요한 역할을 합니다.

각 항목의 의미와 부도 예측에 미치는 영향은 아래와 같습니다.

### 계정 항목 분석:
1. **부채비율**: 총부채를 자기자본으로 나눈 비율. 부채비율이 높을수록 부도 위험이 커질 수 있습니다.
   - **부도 위험**: 높은 부채비율은 부채 상환 부담을 증가시키고, 이를 처리할 수 없는 경우 부도 위험이 증가합니다.
   
2. **자기자본비율**: 자기자본을 총자산으로 나눈 비율. 자기자본 비율이 높으면 기업의 재정적 안정성이 높습니다.
   - **부도 위험**: 자기자본 비율이 낮으면, 부도가 날 확률이 커집니다.
   
3. **차입금의존도**: 총 차입금을 총 자산으로 나눈 비율. 기업이 차입금에 얼마나 의존하는지 나타냅니다.
   - **부도 위험**: 차입금 의존도가 높을수록 상환 부담이 증가하고 부도 위험이 커집니다.
   
4. **유동비율**: 유동자산을 유동부채로 나눈 비율. 유동비율이 높을수록 기업이 단기 부채를 갚을 수 있는 능력이 더 높습니다.
   - **부도 위험**: 유동비율이 낮으면 단기적인 자금 유동성에 문제가 있을 수 있어 부도 위험이 높습니다.
   
5. **비유동비율**: 비유동자산을 총 자산으로 나눈 비율. 비유동자산은 장기적인 자산을 의미합니다.
   - **부도 위험**: 비유동자산 비율이 지나치게 높으면 자산의 유동성이 떨어져 부도 위험을 높일 수 있습니다.
   
6. **매출액영업이익률**: 영업이익을 매출액으로 나눈 비율. 기업이 매출에서 얼마나 효율적으로 이익을 창출하는지 나타냅니다.
   - **부도 위험**: 영업이익률이 낮거나 적자 상태라면 부도 위험이 높습니다.
   
7. **매출액세전순이익률**: 세전순이익을 매출액으로 나눈 비율. 세금 전 이익의 비율을 나타냅니다.
   - **부도 위험**: 세전 순이익률이 낮거나 적자라면 부도 위험이 높습니다.
   
8. **이자보상비율**: 영업이익을 이자비용으로 나눈 비율. 기업이 이자 비용을 얼마나 잘 감당하는지 나타냅니다.
   - **부도 위험**: 이자보상비율이 낮으면 부채 상환에 어려움이 있어 부도 위험이 높습니다.
   
9. **금융비용대매출액**: 금융비용을 매출액으로 나눈 비율. 금융비용이 매출에 비례하여 높은 경우 기업이 많은 금융비용을 지출하고 있음을 나타냅니다.
   - **부도 위험**: 금융비용이 너무 높으면 자금 흐름에 문제가 발생하고 부도 위험이 커집니다.
   
10. **차입금평균이자율**: 차입금에 대한 평균 이자율. 기업이 지불해야 하는 평균 이자율입니다.
    - **부도 위험**: 평균 이자율이 높으면 이자 지불 부담이 커져 부도 위험이 높습니다.
    
11. **인건비대매출액**: 인건비를 매출액으로 나눈 비율. 인건비가 매출에 비례해 많은 경우, 인건비가 경영에 부담을 줄 수 있습니다.
    - **부도 위험**: 인건비가 너무 높으면 경영 효율성이 떨어지고 부도 위험이 커집니다.
    
12. **인건비대영업총비용**: 인건비를 영업총비용으로 나눈 비율. 영업활동에서 인건비가 차지하는 비율을 나타냅니다.
    - **부도 위험**: 인건비 비율이 높으면 인건비 절감이 어려운 상태가 될 수 있어 부도 위험이 증가합니다.
    
13. **매출액증가율**: 매출액의 성장률. 매출이 꾸준히 증가하고 있다면 기업의 성장 가능성이 높다고 볼 수 있습니다.
    - **부도 위험**: 매출액 증가율이 낮거나 감소하고 있으면 부도 위험이 증가합니다.
    
14. **유형자산증가율**: 유형자산의 성장률. 자산의 증가 여부는 기업의 장기적인 안정성에 영향을 미칩니다.
    - **부도 위험**: 유형자산 증가율이 지나치게 높으면 자산의 유동성이 떨어져 부도 위험이 커질 수 있습니다.
    
15. **총자산증가율**: 총자산의 증가율. 자산 증가가 기업의 성장성을 나타냅니다.
    - **부도 위험**: 총자산 증가율이 적거나 감소하면, 재정적 문제가 있을 가능성이 큽니다.

### 부도 예측 변수를 생성하기 위한 가중치 및 가공 방식:
부도 예측을 위한 변수 생성은 각 계정 항목의 **재무 건전성**, **수익성**, **자금 유동성** 등을 기반으로 적절한 가중치를 적용해 종합적인 변수를 도출하는 것입니다.

#### 예시:
1. **부채비율**(Debt Ratio) = 부채 / 자기자본
2. **자기자본비율**(Equity Ratio) = 자기자본 / 총자산
3. **차입금의존도**(Debt Dependency Ratio) = 차입금 / 총자산
4. **유동비율**(Current Ratio) = 유동자산 / 유동부채
5. **매출액영업이익률** = 영업이익 / 매출액
6. **이자보상비율**(Interest Coverage Ratio) = 영업이익 / 이자비용

- **이자보상비율**: 0.10 (이자보상비율이 낮으면 부도 위험 증가)
- **매출액증가율**: 0.10 (매출 감소 시 부도 확률 증가)


```python
import pandas as pd

# 파일 경로 지정 (실제 파일 경로로 변경)
file_path = '/content/drive/MyDrive/기업경영분석지표(2009~,전수조사)_23104224 (1).csv'

# CSV 파일을 읽어오기 (인코딩이 cp949일 때)
df = pd.read_csv(file_path, encoding='cp949')


# 부도 예측에 사용할 계정 항목들
account_columns = [
    '부채비율', '자기자본비율', '차입금의존도', '유동비율', '비유동비율',
    '매출액영업이익률', '매출액세전순이익률', '이자보상비율', '금융비용대매출액',
    '차입금평균이자율', '인건비대매출액', '인건비대영업총비용'
]

# 각 계정 항목에 대한 가중치
weights = {
    '부채비율': 0.15,
    '자기자본비율': 0.20,
    '차입금의존도': 0.10,
    '유동비율': 0.20,
    '비유동비율': 0.10,
    '매출액영업이익률': 0.15,
    '매출액세전순이익률': 0.10,
    '이자보상비율': 0.10,
    '금융비용대매출액': 0.05,
    '차입금평균이자율': 0.05,
    '인건비대매출액': 0.05,
    '인건비대영업총비용': 0.05
}

# 연도 목록 (2009년부터 2023년까지)
years = [str(year) for year in range(2008, 2024)]

# 계정 항목에 해당하는 데이터만 필터링
df_filtered = df[df['계정항목'].isin(account_columns)]

# 부도 예측 점수를 계산하는 함수 (연도별 종합 점수 계산)
def calculate_default_risk_by_year(df_filtered, account_columns, weights, years):
    # 연도별로 부도 예측 점수를 저장할 데이터프레임
    result = pd.DataFrame(columns=['연도', '부도예측점수'])

    # 각 연도에 대해 부도 예측 점수를 계산
    for year in years:
        total_score = 0  # 해당 연도의 부도 예측 점수를 초기화

        # 각 계정 항목에 대해 점수 계산
        for account in account_columns:
            # 해당 계정 항목에 대한 데이터 필터링
            account_data = df_filtered[df_filtered['계정항목'] == account]

            # 해당 계정 항목의 가중치를 가져오기
            weight = weights.get(account, 1)

            # 연도별로 값이 존재하는 경우 점수 계산
            if year in account_data.columns:
                total_score += account_data[year].values[0] * weight

        # 연도별 부도 예측 점수 저장
        result = pd.concat([result, pd.DataFrame({'연도': [year], '부도예측점수': [total_score]})], ignore_index=True)

    return result

# 부도 예측 점수 계산
df_yearly_scores = calculate_default_risk_by_year(df_filtered, account_columns, weights, years)

# 최종 결과를 CSV 파일로 저장
df_yearly_scores.to_csv('부도예측_연도별_점수.csv', index=False)
```

```python
import pandas as pd

# 예시 데이터 (파일 경로를 실제 경로로 변경)
file_path = '/content/drive/MyDrive/기업경영분석지표(2009~,전수조사)_23104224 (1).csv'

# CSV 파일에서 데이터를 읽어오기 (인코딩이 cp949일 때)
df = pd.read_csv(file_path, encoding='cp949')

# 데이터 구조 확인 (계정 항목 및 연도별 데이터 확인)
print(df.head())

# 계정 항목 목록 (부도 예측에 사용할 주요 항목들)
account_columns = [
    '부채비율', '자기자본비율', '차입금의존도', '유동비율', '비유동비율',
    '매출액영업이익률', '매출액세전순이익률', '이자보상비율', '금융비용대매출액',
    '차입금평균이자율', '인건비대매출액', '인건비대영업총비용'
]

# 연도 목록 (2009년부터 2023년까지)
years = [str(year) for year in range(2008, 2024)]

# "계정항목"이 각 계정의 항목을 포함하는 데이터만 필터링
df_filtered = df[df['계정항목'].isin(account_columns)]

# 계정 항목별로 연도별 평균을 계산하는 함수
def calculate_yearly_averages(df, account_columns, years):
    # 계정 항목별로 연도별 평균 계산
    averages = []

    for account in account_columns:
        account_data = df[df['계정항목'] == account]  # 해당 계정 항목에 해당하는 데이터만 선택

        # 계정 항목별 연도별 평균 계산
        avg_values = {}
        for year in years:
            if year in account_data.columns:
                avg_values[year] = account_data[year].mean()  # 연도별 평균 계산

        # 결과를 리스트에 추가
        avg_values['계정항목'] = account
        averages.append(avg_values)

    # DataFrame으로 변환
    averages_df = pd.DataFrame(averages)

    return averages_df

# 계정 항목별 연도별 평균 계산
df_yearly_averages = calculate_yearly_averages(df_filtered, account_columns, years)

# 행과 열을 바꾸기 (transpose)
df_yearly_averages_transposed = df_yearly_averages.set_index('계정항목').transpose()

# 결과 출력 (계정 항목별 연도별 평균, 행과 열이 바뀐 형태)
print(df_yearly_averages_transposed)

# 최종 결과를 CSV로 저장 (행과 열이 바뀐 형태)
df_yearly_averages_transposed.to_csv('계정항목_연도별_평균_전치.csv', index=True)
```
